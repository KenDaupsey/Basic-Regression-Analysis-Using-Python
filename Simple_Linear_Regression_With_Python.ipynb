```python
### Import Necessar Libraries
import pandas as pd
import numpy as np
```

```python
# Load the dataset from GitHub URL
url = "https://raw.githubusercontent.com/KenDaupsey/Simple-Linear--Regression-Analysis-Using-Python/main/hsb2%7Edata.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
### Simple linear regression

import statsmodels.api as sm

# Assuming df is your DataFrame
# Fit linear regression model
X = sm.add_constant(df['read'])  # Add a constant term to the independent variable
y = df['write']

model = sm.OLS(y, X)
result = model.fit()

# Print the regression results
print(result.summary())
```
